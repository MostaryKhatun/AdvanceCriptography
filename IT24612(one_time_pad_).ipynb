```python
import os
import base64

def generate_key(length: int) -> bytes:
    """
    Generate a truly random key of specified length (in bytes).
    Uses os.urandom for cryptographic security.
    """
    return os.urandom(length)

def encrypt(plaintext: str, key: bytes) -> str:
    """
    Encrypt plaintext using XOR between each byte and key byte.
    Returns ciphertext encoded in Base64 for safe storage.
    """
    plaintext_bytes = plaintext.encode('utf-8')
    ciphertext_bytes = bytes([p ^ k for p, k in zip(plaintext_bytes, key)])
    return base64.b64encode(ciphertext_bytes).decode('utf-8')

def decrypt(ciphertext_b64: str, key: bytes) -> str:
    """
    Decrypt Base64-encoded ciphertext using XOR and return plaintext string.
    """
    ciphertext_bytes = base64.b64decode(ciphertext_b64)
    plaintext_bytes = bytes([c ^ k for c, k in zip(ciphertext_bytes, key)])
    return plaintext_bytes.decode('utf-8')

def main():
    print("=== Novel One-Time Pad Encryption System ===")

    plaintext = input("Enter message to encrypt: ").strip()
    key = generate_key(len(plaintext))

    print("\nGenerated Key (Base64 Encoded):")
    print(base64.b64encode(key).decode('utf-8'))

    ciphertext = encrypt(plaintext, key)
    print("\nEncrypted Ciphertext (Base64 Encoded):")
    print(ciphertext)

    decrypted = decrypt(ciphertext, key)
    print("\nDecrypted Message:")
    print(decrypted)

if __name__ == "__main__":
    main()
```

```python
import string
import random

def generate_key():
    """
    Generates a random substitution key.
    """
    letters = list(string.ascii_uppercase)
    shuffled = letters[:]
    random.shuffle(shuffled)
    return dict(zip(letters, shuffled))

def encrypt(plaintext, key):
    """
    Encrypts the plaintext using the substitution key.
    """
    plaintext = plaintext.upper()
    ciphertext = ''.join(key.get(char, char) for char in plaintext)
    return ciphertext

def decrypt(ciphertext, key):
    """
    Decrypts the ciphertext using the substitution key.
    """
    reversed_key = {v: k for k, v in key.items()}
    plaintext = ''.join(reversed_key.get(char, char) for char in ciphertext)
    return plaintext

# Example usage
if __name__ == "__main__":
    plaintext = "HELLO WORLD"
    key = generate_key()

    print("Key:", key)
    ciphertext = encrypt(plaintext, key)
    print("Ciphertext:", ciphertext)

    decrypted_text = decrypt(ciphertext, key)
    print("Decrypted text:", decrypted_text)
```

```python
import os

def generate_key(length):
    """
    Generate a random key of the given length.
    """
    return os.urandom(length)  # Generate truly random bytes

def encrypt(plaintext):
    """
    Encrypt the plaintext using a one-time pad key.
    """
    plaintext_bytes = plaintext.encode('utf-8')
    key = generate_key(len(plaintext_bytes))
    ciphertext = bytes([p ^ k for p, k in zip(plaintext_bytes, key)])
    return ciphertext, key

def decrypt(ciphertext, key):
    """
    Decrypt the ciphertext using the one-time pad key.
    """
    plaintext_bytes = bytes([c ^ k for c, k in zip(ciphertext, key)])
    return plaintext_bytes.decode('utf-8')

# Example usage
if __name__ == "__main__":
    plaintext = "HELLO WORLD"

    # Encrypt
    ciphertext, key = encrypt(plaintext)
    print("Ciphertext (bytes):", ciphertext)
    print("Key (bytes):", key)

    # Decrypt
    decrypted_text = decrypt(ciphertext, key)
    print("Decrypted text:", decrypted_text)
```
